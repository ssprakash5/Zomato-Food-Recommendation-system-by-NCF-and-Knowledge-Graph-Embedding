```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is your Zomato DataFrame

# Extract relevant columns for constructing the knowledge graph
graph_data = df[['name', 'cuisines', 'dish_liked']]

# Initialize a directed graph
knowledge_graph = nx.DiGraph()

# Function to add nodes and edges to the knowledge graph
def add_to_graph(row):
    restaurant_name = row['name']

    # Check for NaN values in 'cuisines' and 'dish_liked'
    cuisines = row['cuisines'].split(', ') if pd.notnull(row['cuisines']) else []
    dish_liked = row['dish_liked'].split(', ') if pd.notnull(row['dish_liked']) else []

    # Add restaurant node
    knowledge_graph.add_node(restaurant_name, type='restaurant')

    # Add cuisine nodes and edges
    for cuisine in cuisines:
        knowledge_graph.add_node(cuisine, type='cuisine')
        knowledge_graph.add_edge(restaurant_name, cuisine, relationship='serves')

    # Add dish nodes and edges
    for dish in dish_liked:
        knowledge_graph.add_node(dish, type='dish')
        knowledge_graph.add_edge(restaurant_name, dish, relationship='likes')

# Apply the function to each row in the DataFrame
graph_data.apply(add_to_graph, axis=1)

# Visualize the graph (optional)
pos = nx.spring_layout(knowledge_graph)
nx.draw(knowledge_graph, pos, with_labels=True, font_size=8, node_size=30)
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split

# Assuming 'df' is your DataFrame
# Extract relevant columns for constructing the knowledge graph
graph_data = df[['name', 'cuisines', 'dish_liked']]

# Create triples for the knowledge graph (restaurant, serves, cuisine) and (restaurant, likes, dish)
triples = []
for index, row in graph_data.iterrows():
    restaurant = row['name']
    cuisines = row['cuisines'].split(', ') if pd.notnull(row['cuisines']) else []
    dish_liked = row['dish_liked'].split(', ') if pd.notnull(row['dish_liked']) else []

    # Add triples for cuisine
    for cuisine in cuisines:
        triples.append((restaurant, 'serves', cuisine))

    # Add triples for dish
    for dish in dish_liked:
        triples.append((restaurant, 'likes', dish))

# Convert triples to PyTorch tensors
entities = set(triple[0] for triple in triples) | set(triple[2] for triple in triples)
relations = set(triple[1] for triple in triples)
num_entities = len(entities)
num_relations = len(relations)

entity_to_id = {entity: idx for idx, entity in enumerate(entities)}
relation_to_id = {relation: idx for idx, relation in enumerate(relations)}

triples_ids = [(entity_to_id[h], relation_to_id[r], entity_to_id[t]) for h, r, t in triples if h in entity_to_id and t in entity_to_id]
triples_tensor = torch.tensor(triples_ids, dtype=torch.long)

# Split the data into training and testing sets
train_triples, test_triples = train_test_split(triples_tensor.numpy(), test_size=0.2, random_state=42)
train_triples = torch.tensor(train_triples, dtype=torch.long)
test_triples = torch.tensor(test_triples, dtype=torch.long)

# Define the HolE model
class HolE(nn.Module):
    def __init__(self, num_entities, num_relations, embedding_dim):
        super(HolE, self).__init__()
        self.embedding_dim = embedding_dim
        self.entity_embeddings = nn.Embedding(num_entities, embedding_dim)
        self.relation_embeddings = nn.Embedding(num_relations, embedding_dim)

    def forward(self, head, relation, tail):
        head_embedding = self.entity_embeddings(head)
        relation_embedding = self.relation_embeddings(relation)
        tail_embedding = self.entity_embeddings(tail)

        # Ensure compatibility by reshaping head_embedding and tail_embedding
        head_embedding = head_embedding.view(-1, 1, self.embedding_dim)
        tail_embedding = tail_embedding.view(-1, 1, self.embedding_dim)

        correlation = torch.sum(head_embedding * tail_embedding, dim=2)
        return torch.sigmoid(correlation * relation_embedding).squeeze()

# Instantiate the HolE model
embedding_dim = 50  # Adjust the embedding dimension as needed
hole_model = HolE(num_entities, num_relations, embedding_dim)

# Define loss function and optimizer
criterion = nn.BCELoss()
optimizer = optim.Adam(hole_model.parameters(), lr=0.001)  # Adjust the learning rate as needed

# Train the HolE model
num_epochs = 50  # Adjust the number of epochs as needed
batch_size = 64  # Adjust batch size as needed

train_dataset = TensorDataset(train_triples[:, 0], train_triples[:, 1], train_triples[:, 2])
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

for epoch in range(num_epochs):
    for batch in train_loader:
        optimizer.zero_grad()
        head, relation, tail = batch
        output = hole_model(head, relation, tail)
        loss = criterion(output, torch.ones_like(output))
        loss.backward()
        optimizer.step()

# Evaluate the HolE model on the test set (you may want to implement metrics such as precision, recall, etc.)
test_dataset = TensorDataset(test_triples[:, 0], test_triples[:, 1], test_triples[:, 2])
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

hole_model.eval()
with torch.no_grad():
    for batch in test_loader:
        head, relation, tail = batch
        output = hole_model(head, relation, tail)
        # Evaluate and print metrics as needed
```

```python
import torch

def compute_mrr(model, test_loader):
    mrr_sum = 0.0
    count = 0

    model.eval()
    with torch.no_grad():
        for batch in test_loader:
            head, relation, tail = batch
            scores = model(head, relation, tail)

            # Sort scores in descending order
            _, indices = torch.sort(scores, descending=True)

            # Find the rank of the correct tail entity (index of tail in sorted list)
            rank = (indices == tail.view(-1, 1)).nonzero()[:, 1] + 1

            # Calculate reciprocal rank and accumulate
            reciprocal_rank = 1.0 / rank.float()
            mrr_sum += reciprocal_rank.sum().item()
            count += len(rank)

    # Calculate mean MRR
    mrr = mrr_sum / count
    return mrr

def compute_hits_at_k(model, test_loader, k=1):
    hits_sum = 0.0
    count = 0

    model.eval()
    with torch.no_grad():
        for batch in test_loader:
            head, relation, tail = batch
            scores = model(head, relation, tail)

            # Sort scores in descending order
            _, indices = torch.sort(scores, descending=True)

            # Check if the correct tail entity is among the top-k predictions
            hits = (indices[:, :k] == tail.view(-1, 1)).any(dim=1).float()
            hits_sum += hits.sum().item()
            count += len(hits)

    # Calculate Hits@k
    hits_at_k = hits_sum / count
    return hits_at_k

def compute_mean_rank(model, test_loader):
    rank_sum = 0.0
    count = 0

    model.eval()
    with torch.no_grad():
        for batch in test_loader:
            head, relation, tail = batch
            scores = model(head, relation, tail)

            # Sort scores in descending order
            _, indices = torch.sort(scores, descending=True)

            # Find the rank of the correct tail entity (index of tail in sorted list)
            rank = (indices == tail.view(-1, 1)).nonzero()[:, 1] + 1

            # Accumulate ranks
            rank_sum += rank.sum().item()
            count += len(rank)

    # Calculate mean rank
    mean_rank = rank_sum / count
    return mean_rank
# Assuming you have a DataLoader named eval_loader for evaluation
mrr = compute_mrr(hole_model, eval_loader)
hits_at_1 = compute_hits_at_k(hole_model, eval_loader, k=1)
mean_rank = compute_mean_rank(hole_model, eval_loader)

print(f'Mean Reciprocal Rank (MRR): {mrr:.4f}')
print(f'Hits@1: {hits_at_1:.4f}')
print(f'Mean Rank: {mean_rank:.4f}')
```

```python
def evaluate_model(model, validation_data):
    model.eval()

    # Extract validation triples
    validation_loader = DataLoader(validation_data, batch_size=batch_size, shuffle=False)

    # Initialize metrics
    mrr_list = []
    hits_at_1_list = []
    rank_list = []

    for batch in validation_loader:
        head, relation, tail = batch
        num_entities = model.entity_embeddings.num_embeddings

        # Create a matrix of size (num_entities x embedding_dim) for each head, relation pair
        head_embedding = model.entity_embeddings(head).unsqueeze(1).expand(-1, num_entities, -1)

        # Expand relation embedding to match the size of head_embedding
        relation_embedding = model.relation_embeddings(relation).unsqueeze(1).expand(-1, num_entities, -1)

        # Expand tail embeddings to match the size of head_embedding
        tail_embedding = model.entity_embeddings.weight.unsqueeze(0).expand_as(head_embedding)

        # Calculate scores
        scores = torch.sum(torch.sigmoid(head_embedding * tail_embedding * relation_embedding), dim=-1)

        # Evaluate metrics for each example in the batch
        for i in range(len(head)):
            positive_score = scores[i, head[i], tail[i]].item()
            ranks = torch.sum(scores[i] > positive_score).item() + 1

            # Calculate metrics
            mrr = 1.0 / ranks
            hits_at_1 = 1 if ranks == 1 else 0

            mrr_list.append(mrr)
            hits_at_1_list.append(hits_at_1)
            rank_list.append(ranks)

    # Calculate overall metrics
    overall_mrr = sum(mrr_list) / len(mrr_list)
    overall_hits_at_1 = sum(hits_at_1_list) / len(hits_at_1_list)
    overall_mean_rank = sum(rank_list) / len(rank_list)

    return overall_mrr, overall_hits_at_1, overall_mean_rank

# Example usage:
validation_data = test_triples  # Replace with your validation data
best_mrr, best_hits_at_1, best_mean_rank = evaluate_model(hole_model, validation_data)
print(f"Best MRR: {best_mrr:.4f}")
print(f"Best Hits@1: {best_hits_at_1:.4f}")
print(f"Best Mean Rank: {best_mean_rank:.4f}")
```
